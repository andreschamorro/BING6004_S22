An open notebook has exactly one interactive session connected to a kernel which will execute code sent by the user and communicate back results. This kernel remains active if the web browser window is closed, and reopening the same notebook from the dashboard will reconnect the web application to the same kernel.

## Jupyter Notebooks, Navigating and Shortcuts

1. Shift-Enter: run cell ** Execute the current cell, show output (if any), and jump to the next cell below. If Shift-Enter is invoked on the last cell, a new code cell will also be created. Note that in the notebook, typing Enter on its own never forces execution, but rather just inserts a new line in the current cell. Shift-Enter is equivalent to clicking the Cell | Run menu item.

2. Ctrl-Enter: run cell in-place ** Execute the current cell as if it were in “terminal mode”, where any output is shown, but the cursor remains in the current cell. The cell’s entire contents are selected after execution, so you can just start typing and only the new input will be in the cell. This is convenient for doing quick experiments in place, or for querying things like filesystem content, without needing to create additional cells that you may not want to be saved in the notebook.

3. Alt-Enter: run cell, insert below ** Executes the current cell, shows the output, and inserts a new cell between the current cell and the cell below (if one exists). (shortcut for the sequence Shift-Enter,Ctrl-m a. (Ctrl-m a adds a new cell above the current one.))

4. Esc and Enter: Command mode and edit mode ** In command mode, you can easily navigate around the notebook using keyboard shortcuts. In edit mode, you can edit text in cells.

```python
import requests
from tqdm import tqdm

def download(url: str, fname: str):
    resp = requests.get(url, stream=True)
    total = int(resp.headers.get('content-length', 0))
    with open(fname, 'wb') as file, tqdm(
        desc=fname,
        total=total,
        unit='iB',
        unit_scale=True,
        unit_divisor=1024,
    ) as bar:
        for data in resp.iter_content(chunk_size=1024):
            size = file.write(data)
            bar.update(size)
```

### Bioinformatics in Python

```python
import random

DNA_BASES = ['T', 'C', 'A', 'G']
AMBIGUITY_CODES = {
    'K': ['G', 'T'],
    'M': ['A', 'C'],
    'R': ['A', 'G'],
    'Y': ['C', 'T'],
    'S': ['C', 'G'],
    'W': ['A', 'T'],
    'B': ['C', 'G', 'T'],
    'V': ['A', 'C', 'G'],
    'H': ['A', 'C', 'T'],
    'D': ['A', 'G', 'T'],
    'X': ['A', 'C', 'G', 'T'],
    'N': ['A', 'C', 'G', 'T']
}

all_characters  = list(DNA_BASES) + sorted(AMBIGUITY_CODES.keys())

def random_sequence(n):
    return ''.join([random.choice(DNA_BASES) for nuc in range(n)])

def remove_ambiguity(read):
    return ''.join([np.random.choice(AMBIGUITY_CODES[n]) if not n in DNA_BASES 
                                                         else n for n in read])
```

```python
def nucolored(seq):
    bcolors = {
        'A': '\033[92m',
        'C': '\033[94m',
        'G': '\033[93m',
        'T': '\033[91m',
        'U': '\033[91m',
        'reset': '\033[0;0m'
    }
    tmp_seq = ""
    for n in seq:
        if n in bcolors:
            tmp_seq += bcolors[n] + n
        else:
            tmp_seq += bcolors['reset'] + n
    return tmp_seq + '\033[0;0m'
```

```python
print(f"DNA + Reverse Complement:\n5' {nucolored(seq_1)} 3'")
#print(f"   {''.join(['|' for c in range(len(seq_1))])}")
print(f"3' {nucolored(reverse_complement(seq_1))} 5'\n")
```

### Calculating a substitution matrix from a pairwise sequence alignment

```python
from Bio import AlignIO
alignments = AlignIO.read("../data/opuntia.aln", "clustal")
print(alignments)
```

However, as they only differ trivially from each other, we arbitrarily choose the first alignment, and count the number of each substitution:

```python
alignment = alignments[0]
from Bio.Align.substitution_matrices import Array
frequency = Array("ACGTN", dims=2)
seq1 = alignments[2].seq
seq2 = alignments[-1].seq
for c1, c2 in zip(seq1, seq2):
    if (c1 in "ACGTN") & (c2 in "ACGTN"):
        frequency[c1, c2] += 1
print(frequency)
```

```python
frequency = alignments.substitutions
print(frequency)
```

We normalize against the total number to find the probability of each substitution, and create a symmetric matrix:

```python
probabilities = frequency / np.sum(frequency)
probabilities = (probabilities + probabilities.transpose()) / 2.0
print(probabilities.format("%.4f"))
```
